```python
complex_df = pd.read_sql_query('''SELECT 
    CASE 
        WHEN AVG(CASE WHEN Year <= 2005 THEN Global_Sales END) > AVG(CASE WHEN Year > 2005 THEN Global_Sales END) 
        THEN 'Average global sales were higher before 2005.'
        WHEN AVG(CASE WHEN Year <= 2005 THEN Global_Sales END) < AVG(CASE WHEN Year > 2005 THEN Global_Sales END)
        THEN 'Average global sales were higher in or after 2005.'
        ELSE 'Both have the same average'
    END AS result
FROM data1202.vgsales;''', conn)
```

```python
pp2005_df = pd.read_sql_query('''SELECT *,
       CASE 
           WHEN Year <= 2005 THEN 'Pre-2005'
           ELSE 'Post-2005'
       END AS Pre_Post2005
FROM data1202.vgsales;''',conn)
```
